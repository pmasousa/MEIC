```python
import os, codecs, re
import nltk
from nltk import grammar, CFG, parse
from nltk.parse.generate import generate
```

```python
grammar1 = CFG.fromstring("""
	S -> NP VP
	NP -> DET N | N PP | DET N PP
  DET -> 'the'
	N -> 'cat'
	N -> 'dog'
	PP -> PREP NP
	VP -> V NP | V PP | V NP PP
	V -> 'chased'
	V -> 'sat'
	PREP -> 'in'
	PREP -> 'on'
	""")

earley1 = parse.EarleyChartParser(grammar1, trace=1)

sentences = ['the dog chased the cat'.split(), 'the cat chased the cat'.split()]
for sentence in sentences:
	chart = earley1.chart_parse(sentence)
	parses = list(chart.parses(grammar1.start()))
	print("\n# eligible assignments:",len(parses))
	for tree in parses:
		print("assignment: %s\n" % tree)
```

```python
grammar2 = CFG.fromstring("""
	F -> SN SV
	SV -> v SN | SV SP
	SP -> prep SN
	SN -> det n | SN SP | Pron
  Pron -> 'I'
	n -> 'man'
	n -> 'hill'
	n -> 'telescope'
	prep -> 'on'
	prep -> 'with'
	v -> 'saw'
	det -> 'a' | 'the'
	""")

sentence2 = 'I saw the man on the hill with a telescope'.split()
earley2 = parse.EarleyChartParser(grammar2, trace=1)
chart2 = earley2.chart_parse(sentence2)
parses2 = list(chart2.parses(grammar2.start()))
print("\n# eligible assignments:",len(parses2))
for i, tree in enumerate(parses2):
	print("\nassignment %d:\n%s" % (i,tree))
```

```python
# Generate possible sentences using the language accepted by a grammar
for sentence in generate(grammar2, depth=5): # try depth = 10 :-)
	print(' '.join(sentence))
```
